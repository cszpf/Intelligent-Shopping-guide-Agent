```python
import re
price_pattern = re.compile(r'(\d+)\.*(\d*)万')
def process_price(s):
    try:
        price = int(s)
        return price
    except ValueError:
        if '万' in s:
            match = re.search(price_pattern,s)
            if match:
                price = int(match.group(1))*10000
                if match.group(2)!='':
                    price += int(match.group(2))*1000
                return price
            else:
                print(s)
                return -1
        else:
            print(s)
            return -1
```

```python
price = data['价格'].apply(process_price)
```
