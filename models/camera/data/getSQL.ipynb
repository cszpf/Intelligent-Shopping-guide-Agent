```python
import re
import numpy as np
price_pattern = re.compile(r'(\d+)\.*(\d*)万')
def process_price(s):
    try:
        price = int(s)
        return price
    except ValueError:
        if '万' in s:
            match = re.search(price_pattern,s)
            if match:
                price = int(match.group(1))*10000
                if match.group(2)!='':
                    
                    price += float('0.'+match.group(2))*10000
                return price
            else:
                #print(s)
                return np.nan
        else:
            #print(s)
            return np.nan
```

```python
price = data['价格'].apply(process_price)
```

```python
brand_list = "佳能 尼康 索尼 富士 宾得 理光 徕卡 松下 奥林巴斯 柯达 三星 哈苏 适马 卡西欧 Insta360 飞思 小米 阿尔帕 爱国者 玛米亚利图 华为 卡尔·蔡司 禄来 永诺 明基 锡恩帝 GE通用电气 米家 OPPO 柏卡 HTC 努比亚 谷歌 宝丽来 联想 Acer 宏碁 小蚁 夏普 QindredCam 萤石 CRAB Google"
brand_list = [b.lower() for b in brand_list.split()]
```

```python
brand = []
for name in data['型号']:
    found = False
    name = name.lower()
    for b in brand_list:
        if b in name:
            brand.append(b)
            found = True
            break
    if not found:
        print(name)
        brand.append("未知")
brand = pd.Series(brand)
```

```python
level_list = '入门 终端 高端 消费 专业 初级 中级 高级 专业 中端 低端 新手级'
level_list = level_list.split()
level = []
for name in data['产品类型']:
    found = False
    if type(name)==float:
        level.append("未知")
        continue
    for b in level_list:
        if b in name:
            level.append(b)
            found = True
            break
    if not found:
        level.append("未知")
level = pd.Series(level)
```

```python
type_list = '卡片 广角 长焦 三防 数码微单 微单 全景 VR相机 单反 胶片 移轴 多功能 家用摄像机'
type_list = [t.lower() for t in type_list.split()]

def extract_type(s):
    if type(s) == float:
        return np.nan
    s = s.lower()
    for t in type_list:
        if t in s:
            return t
    return np.nan

types = data['产品类型'].apply(extract_type)
```

```python
def process_pixel(s):
    if type(s)==float:
        return np.nan
    if '万' in s:
        try:
            s = s.replace('万','')
            s = int(s)
            return s
        except ValueError:
            print(s)
            return np.nan

pixel = data['有效像素'].apply(process_pixel)
```

```python
frame_list = 'APS-C画幅 全画幅 中画幅 m4/3画幅 APS-H画幅 APS画幅'
frame_list = [f.lower() for f in frame_list.split()]
def process_frame(s):
    if type(s)==float:
        return np.nan
    s = s.lower()
    for f in frame_list:
        if f in s:
            return f
    return '半画幅'
frame = data['传感器尺寸'].apply(process_frame)
```

```python
shutter_list = '电磁调控纵走式焦平快门 电子控制垂直焦平面快门 电子控制式金属纵向快门 电子控制纵走式焦平快门 电子快门 电子控制焦平面快门 机械快门 焦平面快门 叶片式快门 b门 旋转式快门 手动快门 脸部快门 电子前帘快门 触摸快门 机械快门 电控焦平面快门'
shutter_list = shutter_list.split()
def shutter_type(s):
    if type(s) == float:
        return np.nan
    s = s.lower()
    for shutter in shutter_list:
        if shutter in s:
            return shutter
    return np.nan

shutter = data['快门类型'].apply(shutter_type)
```

```python
screen_list = '旋转 高清 彩色LCD 触摸 翻转 oled tft彩色液晶 AMOLED TFT液晶 液晶 lcd 三向翻转 宽视角TFT 触控翻折屏'
screen_list = [s.lower() for s in screen_list.split()]

def extract_screen(s):
    if type(s) == float:
        return np.nan
    s = s.lower()
    for screen in screen_list:
        if screen in s:
            return screen
    return np.nan

screen = data['显示屏类型'].apply(extract_screen)
```

```python
sql_data = {'id':data['id'].values,'name':data['型号'],'price':price,'brand':brand,'level':level,'pixel':pixel,'frame':frame,'type':types,'screen':screen,'shutter':shutter,'tags':data['tags'].values}
sql_data = pd.DataFrame(sql_data)
```

```python
from sqlalchemy import Column, String, create_engine
from sqlalchemy.orm import sessionmaker

#%%
engine = create_engine('mysql+mysqlconnector://root@localhost:3306/dialog')
#%%
engine.execute('drop table if exists camera_param')
sql_data.to_sql('camera_param',engine,if_exists = 'replace')
```
